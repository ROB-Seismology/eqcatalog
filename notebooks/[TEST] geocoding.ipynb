```python
import folium
import eqcatalog
```

```python
eq_id = 6625
[eq] = eqcatalog.rob.seismodb.query_local_eq_catalog_by_id(eq_id)
dyfi = eq.get_macroseismic_enquiries()
```

```python
dyfi.get_addresses()[:100]
```

```python
provider = "here"
from here_api_key import app_id, app_code
bbox = [0, 48, 10, 52]
geo_results = dyfi.geocode(provider, bbox=bbox, app_id=app_id, app_code=app_code, max_requests=50)
```

```python
for loc, address in zip(geo_results, dyfi.get_addresses()):
    print(address)
    print(loc)
    print('')
```

```python
m = folium.Map(location=[eq.lat, eq.lon], zoom_start=12)
for loc, address in zip(geo_results, dyfi.get_addresses()):
    if loc:
        folium.Marker([loc[1], loc[0]], popup=address, tooltip=address).add_to(m)
m
```
