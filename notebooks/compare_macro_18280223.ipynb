Compare macroseismic data for 23/02/1828 earthquake in eqmapperdb.Macro18280223 and seismodb.macro_detail

```python
import db.simpledb as simpledb
import eqcatalog
from prettytable import PrettyTable
from eqmapperdb_secrets import database, host, user, passwd
```

```python
def render_html(html):
    from IPython.core.display import display, HTML
    display(HTML(html))
```

```python
eq_date = '18280223'
id_earth = 383
```

```python
DB = simpledb.MySQLDB(database, host, user, passwd)
eqmdb_recs = list(DB.query('Macro%s' % eq_date))
print(len(eqmdb_recs))
```

```python
where_clause = 'id_earth = %d' % id_earth
db_recs = eqcatalog.rob.seismodb.query_seismodb_table('macro_detail', where_clause=where_clause)
print(len(db_recs))
```

```python
db_recs = {rec['id_com']: rec for rec in db_recs}
```

```python
## Communes in eqmapperdb that are not in seismodb
tab = PrettyTable(eqmdb_recs[0].keys())
for rec1 in eqmdb_recs:
    rec2 = db_recs.get(rec1['id_com'])
    if not rec2:
        tab.add_row(rec1.values())
        #print(rec1['id_com'])
render_html(tab.get_html_string())
```

```python
## Records in seismodb/eqmapperdb that have different Imin/Imax
columns = ['id_com', 'Imin (db)', 'Imax (db)', 'Imin (eqmdb)', 'Imax (eqmdb)', 'lastmod']
tab = PrettyTable(columns)
for rec1 in eqmdb_recs:
    rec2 = db_recs.get(rec1['id_com'])
    if rec2:
        if (rec1['Imin'] != rec2['intensity_min'] or rec1['Imax'] != rec2['intensity_max']):
            tab.add_row([rec1['id_com'], rec2['intensity_min'], rec2['intensity_max'],
                         rec1['Imin'], rec1['Imax'], rec2['lastmod']])
render_html(tab.get_html_string())
```
